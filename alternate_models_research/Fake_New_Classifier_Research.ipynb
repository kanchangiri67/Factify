# Hamzab's RoBERTa Fake News Classifier

Model: hamzab/roberta-fake-news-classification

Dataset: Fine-tuned on the Fake and Real News Dataset

Input Format: Requires input in the format: <title> TITLE HERE <content> CONTENT HERE <end>

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

tokenizer = AutoTokenizer.from_pretrained("hamzab/roberta-fake-news-classification")
model = AutoModelForSequenceClassification.from_pretrained("hamzab/roberta-fake-news-classification")

def predict_fake(title, content):
    input_str = f"<title> {title} <content> {content} <end>"
    inputs = tokenizer.encode_plus(input_str, max_length=512, padding="max_length", truncation=True, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
        probs = torch.nn.functional.softmax(outputs.logits, dim=1)[0]
    return {"Fake": probs[0].item(), "Real": probs[1].item()}
```

```python
for topic, content in fake_articles.items():
    news_title = content["title"]
    news_article = content["article"]
    result = predict_fake(news_title, news_article)
    print(result)
```

```python
for topic, content in real_articles.items():
    news_title = content["title"]
    news_article = content["article"]
    result = predict_fake(news_title, news_article)
    print(result)
```
